```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

# Sex and Age relation to 'survived'

```python
tdf = dora.data
tdf['sex'].value_counts()
```

```python
sns.set(style="darkgrid")
sex = tdf['sex'].unique()
count = tdf['sex'].value_counts()
sns.barplot(x=sex, y=count, palette="mako")
```

```python
women = tdf.loc[tdf.sex == 'female']["survived"]
men = tdf.loc[tdf.sex == 'male']["survived"]
rate_w = sum(women)/len(women)
rate_m = sum(men)/len(men)

print("% of women who survived: ", rate_w, "\n% of men who survived: ", rate_m)
```

```python
df = tdf[train['survived']==1]
df.groupby('sex')['age'].plot.hist()
```

Male = orange; Female = blue;
<br>
// <b>As we can see old passengers don't make a great portion of the survivors, and most of the male survivors are between 15 to 40 yo.</b>

```python
df_w = df[df.sex == 'female']
df_w.plot.density(y='age', xlim=[0, 80], figsize=[8, 4])

df_m = df[df.sex == 'male']
df_m.plot.density(y='age', xlim=[0, 80], figsize=[8, 4])
```

// <b>Most of the female survivor have around 25 yo.</b>
<br>
// <b>And most of the male survivor have around 30 yo.</b>

# Class relation to 'survived'

```python
sur = tdf.drop(['Kaggle_Pass_Id', 'name', 'sex', 'age', 'ticket', 'fare', 'embarked', 'boat', 'orig_seq', 'company'], axis=1)
sur.head()
```

```python
corr = sur.corr()
corr.style.background_gradient(cmap = 'mako')
```

```python
clss = sur['pclass'].unique()
cnt = sur['pclass'].value_counts()
sns.barplot(x=clss, y=cnt, palette="mako")
```

// <b>The majority of passengers where on the 3 class.</b>

```python
df_1 = sur[sur.survived == 1]
df_1.pclass.value_counts()
```

```python
df_1.pclass.value_counts(normalize = True)
```

// <b>Most of the survivors were from the 1st class, and a great part of the of victims were from the 3rd.</b>

```python
bar = tdf[tdf['survived']==1]
bar.groupby('sex')['pclass'].plot.hist()
```

```python
bar.groupby('sex')['pclass'].value_counts(normalize = True)
```

// <b>Altough there were more male survivors in number per class, the percentage of women who survived on the 2nd class were higher then men's.</b>
<br>
// <b>Comparing based on the histogram, women survivors are about 50% of men survivors on the 1st class. On the 2nd were about 25%, and on the 3rd class were about 70%.</b>
